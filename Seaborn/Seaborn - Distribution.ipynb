```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_theme() #setting default theme
```

```python
tips = sns.load_dataset('tips')
tips.shape
```

```python
tips.head(10)
```

```python
sns.distplot(tips['total_bill'], bins = 50, kde = False)
plt.show()
```

```python
tips.describe()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips) # default is scatter. Others include reg, hex, and kde
```

```python
sns.pairplot(tips, hue = 'sex')
```

```python
sns.rugplot(tips['total_bill'])
```
