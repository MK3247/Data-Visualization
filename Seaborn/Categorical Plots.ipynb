```python
import seaborn as sns
import numpy as np
sns.set_theme()
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator = np.std) #default is average
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True, dodge = True) #dodge for split
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue = 'sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar') #factorplot to catplot
```
