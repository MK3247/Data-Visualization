## Linear Model Plots

```python
import seaborn as sns 
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
sns.set_theme()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers =['+', '*'], aspect=1, height=8) #height for size
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6, height=8)
```

lmplot stands for linear model plot
