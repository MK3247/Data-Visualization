## Labels 

```python
a = np.array([1,2,3,4])
b = np.array([2,4,6,8])
c = np.array([3,6,9,12])

plt.plot(a,b,a,c,'s:r', lw =3)
plt.title('Burning Calories')
plt.xlabel('days')
plt.ylabel('calories')
plt.show()
```

```python
plt.plot(a,b,a,c,'s:r', lw =3)
plt.title('Burning Calories', loc = 'right')
```

```python
plt.plot(a,b,a,c,'s:r', lw =3)
plt.title('Burning Calories', loc = 'left')
```

## Grid 

```python
plt.plot(a,b,a,c,'s:r', lw =3)
plt.title('Burning Calories', loc = 'right')
plt.grid()
plt.show()
```

```python
plt.plot(a,b,a,c,'s:r', lw =3)
plt.title('Burning Calories', loc = 'right')
plt.grid(axis = 'x')
plt.show()
```

```python
plt.plot(a,b,a,c,'s:r', lw =3)
plt.title('Burning Calories', loc = 'right')
plt.grid(axis = 'y')
plt.show()
```

```python
plt.plot(a,b,a,c,'s:r', lw =3)
plt.title('Burning Calories', loc = 'right')
plt.grid(axis = 'x', ls = '--', lw = 2)
plt.show()
```

## Subplots

```python
a = np.array([1,2,3,4])
b = np.array([2,4,6,8])
c = np.array([3,6,9,12])
d = np.array([4,8,12,16])
plt.subplot(1,3,1)
plt.plot(a,b)
plt.subplot(1,3,2)
plt.plot(c,d)
plt.subplot(1,3,3)
plt.plot(c,a)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(a,b)
plt.title('sales')
plt.subplot(2,1,2)
plt.plot(c,d)
plt.title('manenoz')
plt.show()
```
